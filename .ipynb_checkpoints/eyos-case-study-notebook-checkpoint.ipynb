# Part 1

```python
import numpy as np 
import pandas as pd
import itertools

sales = pd.read_csv('fct_sales.csv', compression='gzip', sep=',')
dims = pd.read_csv('dim_products.csv', compression = 'gzip', sep = ',')
```

```python
#Writing a short algorithm to partition the dataframe into store id, and also chunks of approximately size k

def chunker(k, df):
    
    df_list_id = []
    chunked_df_list = []

    for i,g in df.groupby('store_id'):
        df_list_id.append(pd.DataFrame(g))


    for df_store in df_list_id:
        for  i in range(0,df_store.shape[0],k):
            chunked_df_list.append(df_store[i:i+k])
            
    return chunked_df_list
```

```python
#The algorithm returns a list of dataframe chunks, sorted by store id

chunk_list = chunker(5000, sales)

for i in chunk_list:
    print('Chunk size: {}, Store id: {}'.format(i.shape[0], i.iloc[1,2]))
```

# Part 2

```python
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random
import sklearn

%matplotlib inline
```

```python
#First join the products and stores dataframes

joined = sales.merge(dims, on = 'product_id', how = 'inner')

#Some products on the receipt data are not in the products dataframe. We drop these products during the merge.

grouped = joined.groupby(['store_id','product_category']).sum()[['sales_turnover']]

grouped = grouped.unstack()

#Some stores don't sell products in certain categories, returning NaNs. We replace these with zeros.

grouped = grouped.fillna(0)

#Now we have a dataframe that describes the total sales for each store by product category!

grouped.head()
```

```python
#We attempt to do k-means clustering to uncover different clusters of stores.

from sklearn.cluster import KMeans
from sklearn.model_selection import GridSearchCV

kmeans = GridSearchCV(estimator = KMeans(),
                      param_grid = {'n_clusters' : range(2,8)}
                     )

kmeans.fit(grouped)

grouped['Cluster'] = kmeans.predict(grouped)

print('Best parameters are {}'.format(kmeans.best_params_))
```

```python
#We get the cluster centroids for each cluster

centers = pd.DataFrame(data = kmeans.best_estimator_.cluster_centers_,
                       columns = grouped.columns[1:])

#Replace negative sales numbers with zeros

centers[centers < 0] = 0
```

```python
#Here we do a visualization of the sale distributions by product category for each cluster centroid

fig, ax = plt.subplots(nrows = 2, 
                       ncols = 4, 
                       figsize = (30,24))

fig.suptitle('Store product spreads')

ax = ax.reshape((-1))

for i,axes in enumerate(ax[:7]):
    labels = grouped.droplevel(0, axis = 1).columns[1:]
    axes.pie(centers.iloc[i], labels = labels)
    
plt.show()
```

#### 1)

The stores were clustered into 7 different clusters and each belongs to only one cluster (exclusive grouping). It seems like across the clusters there are two main types of stores: One type obtains their sales revenue mostly through clothing, and the other type through a relatively balanced mix of products.

#### 2b) 

Identify opportunities via 1-1 product associations (if_bought_this_sales_item_id -> likely_to_buy_this_sales_item_id), evaluate and rank the opportunities based on relevant metrics. Code libraries used should only be limited to linear algebra and dataframe manipulation.

#### The analysis attempts to find for each item pair combination, P(Bought A| Bought B). This can be decomposed into P(Bought A and B) / P(Bought B).

```python
#Get the receipt data for each store to study for the pilot

store_balanced = joined[joined['store_id'] == 294]

store_fashion = joined[joined['store_id'] == 23]

#We only need the product id and receipt id
```

```python
#Get a list of the unique products in the store

df = store_balanced

product_list = df['product_id'].unique()
receipt_list = df['receipt_id'].unique()


#Define a function that checks, for one receipt, whether two products are in it

def check_purchased_together(df, receipt_id, product_id_1, product_id_2):
    
    firstcheck = len(df[(df['receipt_id'] == receipt_id) & (df['product_id'] == product_id_1)]) >= 1

    secondcheck = len(df[(df['receipt_id'] == receipt_id) & (df['product_id'] == product_id_2)]) >= 1
    
    return all([firstcheck,secondcheck])


#Define a function that checks, for one receipt, whether one product is in it

def check_purchased_single(df, receipt_id, product_id):
    
    firstcheck = len(df[(df['receipt_id'] == receipt_id) & (df['product_id'] == product_id)]) >= 1
    
    
    return firstcheck


    #Define a function that inputs a pair, and outputs the change in probability between the prior and posterior 

def get_probability_improvement(df, pair):


    purchased_single = 0 #How many times item appears in a receipt
    purchased_double = 0 #How many times item appears in receipt 
    receipt_list = df['receipt_id'].unique()
    
    for r in receipt_list:
        #If receipt contains both items in pair, add one count to purchased_double and purchased_single
        A = pair[0]
        B = pair[1]
        if check_purchased_together(df, 
                                    receipt_id= r,
                                    product_id_1= A,
                                    product_id_2= B):
            purchased_double += 1
            purchased_single += 1
        elif check_purchased_single(df,
                                    receipt_id= r,
                                    product_id= pair[0]):
            purchased_single += 1

    prob_purchased_A_given_purchased_B = purchased_double / purchased_single

    prob_purchased_A = purchased_single / len(receipt_list)

    return abs(prob_purchased_A_given_purchased_B - prob_purchased_A)


#Define a function that checks if the item is bought often enough for consideration

def significant_item(df,product_id):
    p = df[df['product_id'] == product_id].shape[0] / df.shape[0]
    
    if p > 0.01:
        return True
    else:
        return False
    
#Define a function to return pairs of significant items from a store's receipt data:

def return_significant_pairs(df):

    product_list = df['product_id'].unique()
    receipt_list = df['receipt_id'].unique()

    items = [i for i in product_list if significant_item(df,i)]
    pairs = list(itertools.combinations(items, 2))
    
    return pairs
```

```python
#For each pair, we go through every receipt

#Define a function that inputs a pair, and outputs the change in probability between the prior and posterior 

def get_probability_improvement(df,pair):


    purchased_single = 1 #How many times item appears in a receipt
    purchased_double = 1 #How many times item appears in receipt 

    for r in receipt_list:
        #If receipt contains both items in pair, add one count to purchased_double and purchased_single
        A = pair[0]
        B = pair[1]
        if check_purchased_together(df, 
                                    receipt_id= r,
                                    product_id_1= A,
                                    product_id_2= B):
            purchased_double += 1
            purchased_single += 1
        elif check_purchased_single(df,
                                    receipt_id= r,
                                    product_id= pair[0]):
            purchased_single += 1

    prob_purchased_A_given_purchased_B = purchased_double / purchased_single

    prob_purchased_A = purchased_single / len(receipt_list)

    return abs(prob_purchased_A_given_purchased_B - prob_purchased_A)
```

#### Now that all functions are defined, we calculate the improvement in probability points for both stores:

```python
#Now that all functions are defined, we calculate the improvement in probability points for both stores:

sigpairs_balanced = return_significant_pairs(store_balanced)
sigpairs_fashion = return_significant_pairs(store_fashion)
```

```python
%%time

balanced_improvement = 0

for i in sigpairs_balanced:
    improvement = get_probability_improvement(store_balanced,i)
    balanced_improvement += improvement   
    
```

```python
balanced_improvement
```

```python
fashion_improvement = 0

for i in sigpairs_fashion:
    improvement = get_probability_improvement(store_fashion,i)
    fashion_improvement += improvement
```

#### We can see that the improvement in probability points for the "balanced" store (Cluster 1) is 9.567, significantly higher than the improvement for the "fashion" store (Cluster 4) at 0.99. 

Also the improvement for the clothes-focused store came only from one item-pair, as no other items were significant enough to be considered in the algorithm (item must be responsible for at least 1% of store sales).

#### 2c.

There is definitely a larger opportunity to increase sales through analysing 1-1 product associations in the store that obtains revenue through a balanced distribution of goods, compared to the store that obtains revenue through mostly clothing.

This might be because clothing items tend to have many different variations of the same product type (50 different kinds of shirts in a store), so finding significant associations between specific designs may be unlikely. A better analysis in this type of store might be to compare 1-1 product type associations (whether buying a shirt makes it more likely to buy a pair of shorts) instead of looking for associations between the individual products themselves.
